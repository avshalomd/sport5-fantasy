# User Interface

### General
* NUM_ROUNDS - number of rounds to generate data on. NUM_ROUNDS <= number of rounds raw data in this repo.

```python
NUM_ROUNDS = 7
```

### Teams
* BO_WEIGHT, PR_WEIGHT, FRM_WEIGHT - weights of betting odds, mean power rank and form in last 6 matches respectively. The sum of all 3 weights should be equal to 1.
* WEIGHTED_SUM_POWER - power of the weighted sum of team's betting odds, mean power rank and form in last 6 matches.
* WEIGHTED_SUM_DIFF_WEIGHT - weight of difference between 2 teams weighted sum (betting odds, mean power rank and form in last 6 matches). when there is a considerable difference, the better team gets a "bonus" by this weight.
* TOURNAMENT_TOUGHNESS - value correcting the bais in "Goals For" and "Goals Against" from easier tournaments. the assumption is that the static data of the teams are taken from easier tournaments.

```python
BO_WEIGHT = 0.4                # [0.0, 1.0] where sum(BO_WEIGHT,PR_WEIGHT,FRM_WEIGHT)=1
PR_WEIGHT = 0.25               # [0.0, 1.0]
FRM_WEIGHT = 0.35              # [0.0, 1.0]
WEIGHTED_SUM_POWER = 1.4       # >= 1.0
WEIGHTED_SUM_DIFF_WEIGHT = 1.3 # [1.0, 2.0]
TOURNAMENT_TOUGHNESS = 0.2     # [0.0, 1.0] where 0 is easiest and 1.0 is toughest
```

### Players
* PENALTY_PER_MATCH - number of predicted penalties per game
* SCORE_PENALTY_ODDS - the odds for a penalty taker player to score a penalty

```python
PENALTY_PER_MATCH = 0.33  # > 0.0
SCORE_PENALTY_ODDS = 0.60 # [0.0, 1.0]
SAVE_PENALTY_ODDS = 0.75  # [0.0, 1.0], Note: odds out of penalty misses
```

# Predict matches outcome using teams statistics
## Teams attributes
1. name_h - team hebraw name (str)
2. name_e - team english name (str)
3. theguardian_pr - power ranking extracted from "The Guardian" website. (int: 1->24)
4. fifa_ranking - ordered list of teams by fifa ranking. (int: 1->24)
5. reddit1_pr - power ranking extracted from a sub-reddit. (int: 1->24) 
6. reddit2_pr - another power ranking extracted from a sub-reddit. (int: 1->24)
7. my_pr - my personal power ranking of teams. (int: 1->24)
8. bets_odds - average betting odds (for winning the tournament) extracted from oddschecker. (float: >0 )
9. form_last6 - from of the team in the last 6 games (from left to right), W-Win, D-Draw, L-Lose. (str)
10. gfpg - Goals For per Game - qualifiers round. (float: >=0)
10. gapg - Goals Against per Game - qualifiers round. (float: >=0)

```python
teams_df = pd.read_csv('data/teams.csv', encoding='utf-8')
teams_df.head()
```

## Calculate teams "strength" using their attributes
* I'll create a general PR (power rank) score for each team based on combination of other power ranks sources, betting odds and form.
* based on the above and considering GFpG (Goals For per Game) and GApG (Goals Against per Game) i'll try to estimate the score for each match.
* based on the estimated score of each match and the scoring rules of the game, i'll try to estimate how much game points each player "worth" every round.

* form - accumulated points according to form of last 6 matches. normalized to (0,1). higher is better.

```python
teams_df['form'] =  3 * teams_df['form_last6'].str.count('W') +  teams_df['form_last6'].str.count('D')
minimum = teams_df['form'].min()
maximum = teams_df['form'].max()
teams_df['form'] = (teams_df['form'] - minimum)/(maximum - minimum)
teams_df.drop('form_last6', axis='columns', inplace=True)
```

* power_rank_avg - average of 5 different power ranking (detailed above). normalized to (0,1). higher is better.

```python
power_rank_columns = ["theguardian_pr", "fifa_ranking", "reddit1_pr", "reddit2_pr", "my_pr"]
teams_df['power_rank_avg'] = teams_df[power_rank_columns].mean(axis=1)
minimum = teams_df['power_rank_avg'].min()
maximum = teams_df['power_rank_avg'].max()
teams_df['power_rank_avg'] = (teams_df['power_rank_avg'] - minimum)/(maximum - minimum)
teams_df['power_rank_avg'] = 1 - teams_df['power_rank_avg']
teams_df.drop(power_rank_columns, axis='columns', inplace=True)
```

* bets_odds - average betting odds (detailed above). normalized to (0,1). higher is better.

```python
minimum = teams_df['bets_odds'].min()
maximum = teams_df['bets_odds'].max()
teams_df['bets_odds'] = (teams_df['bets_odds'] - minimum)/(maximum - minimum)
teams_df['bets_odds'] = 1 - teams_df['bets_odds']
```

* weighted_sum_pr - power rank that is a wighted sum of form, power_rank_avg and bets_odds. note that the weight of bet odds is significantly higher. it's probably the most accurate representation of the ranking between teams.

```python
w_bets_odds = BO_WEIGHT
w_power_rank_avg = PR_WEIGHT
w_form = FRM_WEIGHT
w_power = WEIGHTED_SUM_POWER
teams_df['weighted_sum'] = (w_bets_odds * teams_df['bets_odds']) + (w_power_rank_avg * teams_df['power_rank_avg']) + (w_form * teams_df['form'])
teams_df['weighted_sum'] = np.power(teams_df['weighted_sum'], w_power)
```

* gfpg, gapg - Goals For/Against per Game. I tweeked it to represent more accurately the harder match-ups in the tournament (relitivly to qulifiers, where gfpg and gapg were recorded). 

```python
toughness_weight = TOURNAMENT_TOUGHNESS
teams_df['gfpg'] = (1 - toughness_weight) * teams_df['gfpg']
teams_df['gapg'] = (1 + toughness_weight) * teams_df['gapg']
```

```python
cols = ['bets_odds', 'form', 'power_rank_avg', 'weighted_sum']
teams_df[cols] = teams_df[cols].round(3)
teams_df.head()
```

## Load mtaches data

```python
matches_columns_renames = {"תאריך": "date",
                           "שעה": "hour",
                           "קבוצה מארחת": "home_team",
                           "תוצאה": "score",
                           "קבוצה אורחת": "away_team",
                           "סקירה": "review"}

matches_by_round_list = []
for i in range(NUM_ROUNDS):
    raw_data_path = f'data/raw/2021-06-06-r{i+1}.html'
    data = pd.read_html(raw_data_path)
    matches_df = data[0].rename(columns=matches_columns_renames)
    matches_df['date'] = pd.to_datetime(matches_df['date'], format="%d.%m.%Y")
    del matches_df['hour']
    del matches_df['review']
    matches_by_round_list.append(matches_df)
```

```python
pr_diff_wight = WEIGHTED_SUM_DIFF_WEIGHT

def calc_GF(x):
    numenator = x.gfpg + x.opponent_gapg
    denuminator = 2 - pr_diff_wight * x.pr_diff
    return numenator / denuminator

def calc_GA(x):
    numenator = x.gapg + x.opponent_gfpg
    denuminator = 2 + pr_diff_wight * x.pr_diff
    return numenator / denuminator
```

```python
matches_outcome_by_round_list = []
opponents_by_round_list = []
for i in range(NUM_ROUNDS):
    opponents1_df = (matches_by_round_list[i])[['home_team', 'away_team']].set_index('home_team')
    opponents2_df = (matches_by_round_list[i])[['home_team', 'away_team']].set_index('away_team')
    opponents_df = pd.concat([opponents1_df, opponents2_df])
    opponents_df['opponent'] = opponents_df.fillna('').sum(axis=1)
    opponents_df.drop(columns=['home_team', 'away_team'], inplace=True)
    opponents_by_round_list.append(opponents_df)

    matches_df = opponents_df.merge(teams_df, left_index=True, right_on='name_h').set_index('name_h')
    matches_df.drop(columns=['name_e', 'bets_odds', 'power_rank_avg', 'form'], inplace=True)
    matches_df.rename(columns={'weighted_sum': 'pr'}, inplace=True)
    matches_df['opponent_pr'] = matches_df.apply(lambda row: matches_df.loc[row['opponent']].get('pr'), axis=1)
    matches_df['opponent_gfpg'] = matches_df.apply(lambda row: matches_df.loc[row['opponent']].get('gfpg'), axis=1)
    matches_df['opponent_gapg'] = matches_df.apply(lambda row: matches_df.loc[row['opponent']].get('gapg'), axis=1)
    matches_df['pr_diff'] = matches_df['pr'] - matches_df['opponent_pr']
    matches_df.drop(columns=['pr', 'opponent_pr'], inplace=True)

    matches_df['pred_GF'] = matches_df.apply(lambda row: calc_GF(row), axis=1)
    matches_df['pred_GA'] = matches_df.apply(lambda row: calc_GA(row), axis=1)
    matches_df['pred_score'] = matches_df.apply(lambda row: (np.floor(row.pred_GF).astype(int), np.floor(row.pred_GA).astype(int)), axis=1)
    matches_outcome_by_round_list.append(matches_df)
```

## Load and claculate players statistics
*based on whoscored.com data of last 2 tournaments

```python
teams_names = teams_df['name_e'].values
teams_stats_df_dict = {}
for team in teams_names:
#     print(team)
    players_stats_by_team = pd.read_html(f'data/raw/teams/{team}/r6.html')
    curr_team_df = (players_stats_by_team[1])[['Player.1', 'Apps', 'Mins', 'Goals', 'Assists', 'SpG', 'Rating']]
    curr_team_df.rename(columns={"Player.1": "Player"}, inplace=True)
    curr_team_df = curr_team_df.replace('-', 0)
    curr_team_df['Player'] = curr_team_df['Player'].apply(lambda x: (x.split(',')[0])[:-3])
    curr_team_df['Apps'] = curr_team_df['Apps'].apply(lambda x: (x.split('(')[0]))
    teams_stats_df_dict[f'{team}'] = curr_team_df
    
# teams_stats_df_dict2 = {}
# for team in teams_names:
# #     print(team)
#     players_stats_by_team = pd.read_html(f'data/raw/teams/{team}/r11.html')
#     curr_team_df = (players_stats_by_team[1])[['Player.1', 'Apps', 'Mins', 'Goals', 'Assists', 'SpG', 'Rating']]
#     curr_team_df.rename(columns={"Player.1": "Player"}, inplace=True)
#     curr_team_df = curr_team_df.replace('-', 0)
#     curr_team_df['Player'] = curr_team_df['Player'].apply(lambda x: (x.split(',')[0])[:-3])
#     curr_team_df['Apps'] = curr_team_df['Apps'].apply(lambda x: (x.split('(')[0]))
#     teams_stats_df_dict2[f'{team}'] = curr_team_df
```

```python
convert_dict = {'Player': str,
                'Apps': int,
                'Mins': int,
                'Goals': int,
                'Assists': int,
                'SpG': float,
                'Rating': float}

teams_stats_merged_df_dict = {}
for team in teams_names:
#     df = teams_stats_df_dict[team].append(teams_stats_df_dict2[team])
    df = teams_stats_df_dict[team]
    df = df.astype(convert_dict)
    grouped = df.groupby('Player').agg({'Apps': 'sum',
                                    'Mins': 'sum',
                                    'Goals': 'sum',
                                    'Assists': 'sum',
                                    'SpG': 'mean',
                                    'Rating': 'mean'})
    apps_max = grouped['Apps'].max()
    mins_max = grouped['Mins'].max()
    goals_sum = grouped['Goals'].sum()
    assists_sum = grouped['Assists'].sum()
    grouped['Apps_percentage'] = grouped['Apps'] / apps_max
    grouped['Mins_percentage'] = grouped['Mins'] / mins_max
    grouped['Goals_percentage'] = grouped['Goals'] / goals_sum
    grouped['Assists_percentage'] = grouped['Assists'] / assists_sum
    grouped['GnA_percentage'] = (grouped['Goals'] + grouped['Assists']) / (goals_sum + assists_sum)
    teams_stats_merged_df_dict[team] = grouped
```

```python
players_stats_df = pd.concat(list(teams_stats_merged_df_dict.values()))
players_stats_df.head()
```

## Calculate predicted points baset on players statistics and match outcome

```python
# TODO: remove is_starter cond. let filter handle it
penalty_per_match = PENALTY_PER_MATCH
score_penalty_odds = SCORE_PENALTY_ODDS

penalty_taker_bonus = penalty_per_match * score_penalty_odds
penalty_taker_penalty = 4 * penalty_per_match * (1 - score_penalty_odds)
penalty_save_bonus = 4 * penalty_per_match * (1 - score_penalty_odds) * SAVE_PENALTY_ODDS

def goals_against_bonus(goals_num):
    if goals_num == 0:
        return 4
    else:
        return 1 - goals_num

def calc_gk_score(x):
    return x.is_starter * (2 + goals_against_bonus(np.floor(x.pred_GA)) + penalty_save_bonus)

def calc_d_score(x):
    return x.is_starter * (2 + goals_against_bonus(np.floor(x.pred_GA)) + 3*x['Assists_percentage']*x.pred_GF + 6*x['Goals_percentage']*x.pred_GF + x.is_penalty_taker*(6*penalty_taker_bonus - penalty_taker_penalty))

def calc_m_score(x):
    return x.is_starter * (2 + 3*x['Assists_percentage']*x.pred_GF + 5*x['Goals_percentage']*x.pred_GF + x.is_penalty_taker*(5*penalty_taker_bonus - penalty_taker_penalty))

def calc_fw_score(x):
    return x.is_starter * (2 + 3*x['Assists_percentage']*x.pred_GF + 4*x['Goals_percentage']*x.pred_GF + x.is_penalty_taker*(4*penalty_taker_bonus - penalty_taker_penalty))
```

```python
gk_predicted_scores = all_rounds_players_df[all_rounds_players_df['position'] == 'GK']
gk_predicted_scores['pred_points'] = gk_predicted_scores.apply(lambda row: calc_gk_score(row), axis=1)

d_predicted_scores = all_rounds_players_df[all_rounds_players_df['position'] == 'D']
d_predicted_scores['pred_points'] = d_predicted_scores.apply(lambda row: calc_d_score(row), axis=1)

m_predicted_scores = all_rounds_players_df[all_rounds_players_df['position'] == 'M']
m_predicted_scores['pred_points'] = m_predicted_scores.apply(lambda row: calc_m_score(row), axis=1)

fw_predicted_scores = all_rounds_players_df[all_rounds_players_df['position'] == 'FW']
fw_predicted_scores['pred_points'] = fw_predicted_scores.apply(lambda row: calc_fw_score(row), axis=1)

players_predicted_score_df = pd.concat([gk_predicted_scores, d_predicted_scores, m_predicted_scores, fw_predicted_scores])
players_predicted_score_df.head()
```
